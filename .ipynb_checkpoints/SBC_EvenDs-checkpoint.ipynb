```python
import pickle

a_file = open("IANB_dic.pkl", "rb")
IANBC_dic = pickle.load(a_file)
a_file.close()
```

```python
images_name = ['lena', 'cam', 'man', 'liv', 'gol', 'zelda']

import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

font = font_manager.FontProperties(
    family='Times New Roman',
    weight='bold',
    style='normal', size= 30)

intervals = range(2,129)
even_ds = range(2,65,2)
for k in [0,1,2,3,4,5]:
    hcorrs0 = IANBC_dic[images_name[k]]['hcorrs0']
    hcorrs1 = IANBC_dic[images_name[k]]['hcorrs1']
    hcorrs = (hcorrs0 + hcorrs1) /2
    h64even = hcorrs[0:64:2]
    h43even = hcorrs[0:42:2]
    
    vcorrs0 = IANBC_dic[images_name[k]]['vcorrs0']
    vcorrs1 = IANBC_dic[images_name[k]]['vcorrs0']
    vcorrs = (vcorrs0 + vcorrs1) /2
    v64even = vcorrs[0:64:2]
    v43even = vcorrs[0:42:2]
    
    dcorrs0 = IANBC_dic[images_name[k]]['dcorrs0']
    dcorrs1 = IANBC_dic[images_name[k]]['dcorrs0']
    dcorrs = (dcorrs0 + dcorrs1) /2
    d64even = dcorrs[0:64:2]
    d43even = dcorrs[0:42:2]
    
    # Correlations
    plt.figure(figsize =(20, 10) )
    plt.plot(even_ds, h64even, marker='o', label = 'HCorr')
    plt.plot(even_ds, v64even, marker='o', label = 'VCorr')
    plt.plot(even_ds, d64even, marker='o', label = 'DCorr')

    plt.grid(True, which='major', linestyle='--' )
    plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
    plt.minorticks_on()

    plt.xlabel('d', style='italic', fontname="Times New Roman", fontsize = 40)
    plt.ylabel('Correlations', fontname="Times New Roman", fontsize = 40)

    plt.xticks(fontsize = 40, fontname="Times New Roman")
    plt.yticks(fontsize = 40, fontname="Times New Roman")

    plt.rcParams['axes.axisbelow'] = True
    plt.legend(prop=font)

    graph_name = images_name[k]+'_corr.png'
    plt.savefig(graph_name, format='png', dpi=300, bbox_inches = 'tight')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager


def d(t, m):
    b = 8
    return np.floor((2**(b-1)-1-m)/(t+1))
   
m = np.arange(2,128, 2)

a = plt.figure(figsize =(20, 10))
dif = {}
dtm = {}
for t in range(1, 6):
    plt.scatter(m, d(t,m), s = 20, label = f't={t}')
    dif[t] = d(t,m) - np.floor(m/1)
    dtm[t] = d(t,m)
    
plt.scatter(m, np.floor(m/1), label = f't={1} same $L_i$')


plt.xticks(fontsize = 40, fontname="Times New Roman")
plt.yticks(fontsize = 40, fontname="Times New Roman")
plt.grid(True, which='major', linestyle='--' )
plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
plt.minorticks_on()

plt.xlabel('p', style='italic', fontname="Times New Roman", fontsize = 40)
plt.ylabel('$\mathregular{d_t}$', style='italic', fontname="Times New Roman", fontsize = 40)


font = font_manager.FontProperties(
    family='Times New Roman',
    weight='bold',
    style='normal', size= 30)
plt.legend(prop=font)

graph_name ='dpt.png'
plt.savefig(graph_name, format='png', dpi=300, bbox_inches = 'tight')
plt.show()
```
