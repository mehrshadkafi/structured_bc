```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
```

```python
def list_creation(b, d):
    List = {}
    nis = []
    for i in range(0, int(d/2)):
        ni = int((2**(b-1)-2*i-1)/d)
        nis.append(ni) 
        smaller_list = []
        for k in range(0, ni+1):
            smaller_list.append(2**(b-1)+2*i+k*d)

        List[i] = smaller_list
        
    return List    
```

```python
def complete_list_creation(b, d):
    List = list_creation(b, d)
    complete_list = {}
    for i in List.keys():
        complete_smaller_list = []
        for j in List[i]:
            k = j + 1
            complete_smaller_list.append(j)
            complete_smaller_list.append(k)

        complete_list[i] = complete_smaller_list
    return complete_list
```

```python
def f2_method1(b, d):
    complete_list = complete_list_creation(b, d)
    combined_list1 = []
    for i in complete_list.keys():
        combined_list1 += complete_list[i]
    return combined_list1

def f2_method2(b, d):
    complete_list = complete_list_creation(b, d)
    combined_list2 = []
    for i in range(0, int(d/4)):
        j = complete_list[i] + complete_list[i + int(d/4)]
        combined_list2 += j
    return combined_list2
```

```python
def min_distance_method1(b, d, t):
       
    f1s = list(range(0,int(2**(b-1))))
    
#     List = list_creation(b, d)
#     complete_list = complete_list_creation(List)
    '''Method of listing matters here.'''
    f2s = f2_method1(b, d)
    
    distances = []
    for y in range(0, 2**b - t):
        yt = y + t

        if y in f1s and yt in f1s:
            fy = f2s[f1s.index(y)]
            fyt = f2s[f1s.index(yt)]

        if y in f2s and yt in f2s:
            fy = f1s[f2s.index(y)]
            fyt = f1s[f2s.index(yt)]

        if y in f1s and yt in f2s:
            fy = f2s[f1s.index(y)]
            fyt = f1s[f2s.index(yt)]

        distances.append(abs(fyt - fy))
        min_distance = min(distances)
        
    return min_distance#, distances
```

```python
def min_distance_method2(b, d, t):
       
    f1s = list(range(0,int(2**(b-1))))
    
#     List = list_creation(b, d)
#     complete_list = complete_list_creation(List)
    '''Method of listing matters here.'''
    f2s = f2_method2(b, d)
    
    distances = []
    for y in range(0, 2**b - t):
        yt = y + t

        if y in f1s and yt in f1s:
            fy = f2s[f1s.index(y)]
            fyt = f2s[f1s.index(yt)]

        if y in f2s and yt in f2s:
            fy = f1s[f2s.index(y)]
            fyt = f1s[f2s.index(yt)]

        if y in f1s and yt in f2s:
            fy = f2s[f1s.index(y)]
            fyt = f1s[f2s.index(yt)]

        distances.append(abs(fyt - fy))
        min_distance = min(distances)
    
    '''All distances can be returned here.'''        
    return min_distance#, distances
```

```python
b = 8
min_dt1 = {}
min_dt2 = {}
for t in range(2, 10, 2):
    min_ds1 = []
    min_ds2 = []
    for d in range (4, 2**(b-2), 4):
        min_ds1.append(min_distance_method1(b, d, t))
        '''For method2 d must be multiple of 4.'''
        min_ds2.append(min_distance_method2(b, d, t))
        
    min_dt1[t] = min_ds1
    min_dt2[t] = min_ds2    
```

```python
'''Different ts for method1 of listing'''

ds = range (4, 2**(b-2),4)
ts = range (2, 10, 2)
plt.figure(figsize =(20, 10))
for t in ts: 
    plt.plot(ds, min_dt1[t], marker='o', label = f't = {t}')

    plt.xticks(ds,fontsize = 40, fontname="Times New Roman")
    plt.yticks(fontsize = 40, fontname="Times New Roman")
    plt.grid(True, which='major', linestyle='--' )
    plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
    plt.minorticks_on()

    plt.xlabel('d', style='italic', fontname="Times New Roman", fontsize = 40)
    plt.ylabel(r'$\delta(d)$', style='italic', fontname="Times New Roman", fontsize = 30)
    
    font = font_manager.FontProperties(
    family='Times New Roman',
    weight='bold',
    style='normal', size= 30)
    plt.legend(prop=font)
```
