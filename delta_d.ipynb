```python
b = 8
ds = range(2, 2**(b-2)+1, 2)

def delta_d(b, d):
    if d <= 2**(b/2):
        return d
    else:
        return 2 * int((2**(b-1)+1)/d)

delta_ds = [delta_d(b, d) for d in ds]
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from scipy.interpolate import interp1d

plt.figure(figsize =(20, 10) )
plt.plot(ds, delta_ds, marker='o')

delta_d_margin = 10
delta_d_margins = [d*0+delta_d_margin for d in ds]
plt.plot(ds, delta_d_margins)

# f2 = interp1d(ds, delta_ds, kind='quadratic')
# plt.plot(ds, f2(ds))


plt.xticks(fontsize = 40, fontname="Times New Roman")
plt.yticks(fontsize = 40, fontname="Times New Roman")
plt.grid(True, which='major', linestyle='--' )
plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
plt.minorticks_on()

plt.xlabel('d', style='italic', fontname="Times New Roman", fontsize = 40)
plt.ylabel(r'$\delta(d)$', style='italic', fontname="Times New Roman", fontsize = 30)

fig_name = 'delta_d.png'
plt.savefig(fig_name, format='png', dpi=300, bbox_inches = 'tight')
plt.show()
```
