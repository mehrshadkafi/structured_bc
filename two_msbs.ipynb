```python
import cv2
import numpy as np
```

```python
b = 8

U00 = range(0, 2**(b-2))
U01 = range(2**(b-2), 2**(b-1))
U10 = range(2**(b-1), 3 * 2**(b-2))
U11 = range(3 * 2**(b-2), 2**b)

edge1 = 2**(b-2)
edge2 = 2**(b-1)
edge3 = 3 * 2**(b-2)
```

```python
def fellow_finder(value):
    category = value % 4

    if category == 0:
        if value in U00:
            f00 = value
            f01 = value + 64
            f10 = value + 128
            f11 = value + 192

        if value in U01:
            f00 = value - 64
            f01 = value 
            f10 = value + 64
            f11 = value + 128

        if value in U10:
            f00 = value - 128
            f01 = value - 64
            f10 = value 
            f11 = value + 64

        if value in U11:
            f00 = value - 192
            f01 = value - 128
            f10 = value - 64
            f11 = value

    if category == 1:
        if value in U00:
            f00 = value + edge2
            f01 = value + edge3
            f10 = value
            f11 = value + edge1

        if value in U01:
            f00 = value + edge1
            f01 = value + edge2
            f10 = value - edge1
            f11 = value

        if value in U10:
            f00 = value
            f01 = value + edge1
            f10 = value - edge2
            f11 = value - edge1

        if value in U11:
            f00 = value - edge1
            f01 = value
            f10 = value - edge3
            f11 = value - edge2

    if category == 2:
        if value in U00:
            f00 = value + edge1
            f01 = value
            f10 = value + edge3
            f11 = value + edge2

        if value in U01:
            f00 = value 
            f01 = value - edge1
            f10 = value + edge2
            f11 = value + edge1

        if value in U10:
            f00 = value - edge1
            f01 = value - edge2
            f10 = value + edge1
            f11 = value 

        if value in U11:
            f00 = value - edge2
            f01 = value - edge3
            f10 = value
            f11 = value - edge1

    if category == 3:
        if value in U00:
            f00 = value + edge3
            f01 = value + edge2
            f10 = value + edge1
            f11 = value 

        if value in U01:
            f00 = value + edge2
            f01 = value + edge1
            f10 = value 
            f11 = value - edge1

        if value in U10:
            f00 = value + edge1
            f01 = value 
            f10 = value - edge1
            f11 = value - edge2

        if value in U11:
            f00 = value
            f01 = value - edge1
            f10 = value - edge2
            f11 = value - edge3
            
    return f00, f01, f10, f11
```

```python
'''Reconstructions'''
def reconstruction_2msb (image_name):
    '''Images_name should be like: 'lena', 'cam', 'man', 'liv', 'gol', or 'zelda' 
    in tif format.'''

    img_file = image_name +'.tif'
    intensities = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
    imf = np.float32(intensities)
    height = np.shape(imf)[0]
    width = np.shape(imf)[1]

    eve_msb00 = np.zeros([np.shape(imf)[0],np.shape(imf)[1]])
    eve_msb01 = np.zeros([np.shape(imf)[0],np.shape(imf)[1]])
    eve_msb10 = np.zeros([np.shape(imf)[0],np.shape(imf)[1]])
    eve_msb11 = np.zeros([np.shape(imf)[0],np.shape(imf)[1]])
    for i in range(np.shape(imf)[0]):
        for j in range(np.shape(imf)[1]):

            fellows = fellow_finder(imf[i,j])

            eve_msb00[i,j] = fellows[0]
            eve_msb01[i,j] = fellows[1]
            eve_msb10[i,j] = fellows[2]
            eve_msb11[i,j] = fellows[3]

    eve_msb00_uint8 = eve_msb00.astype(np.uint8)
    filename = image_name+'_msb00.png'
    cv2.imwrite(filename, eve_msb00_uint8)

    eve_msb01_uint8 = eve_msb01.astype(np.uint8)
    filename = image_name+'_msb01.png'
    cv2.imwrite(filename, eve_msb01_uint8)

    eve_msb10_uint8 = eve_msb10.astype(np.uint8)
    filename = image_name+'_msb10.png'
    cv2.imwrite(filename, eve_msb10_uint8)

    eve_msb11_uint8 = eve_msb11.astype(np.uint8)
    filename = image_name+'_msb11.png'
    cv2.imwrite(filename, eve_msb11_uint8)
```

```python
for image_name in ['cam', 'man', 'liv', 'gol', 'zelda']:
# for image_name in ['lena']:
    reconstruction_2msb(image_name)
```
