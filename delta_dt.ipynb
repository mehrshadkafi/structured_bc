```python
b = 8
B = 2**(b-1)
ds = range(2, 2**(b-2)+1, 2)

t = 4
delta_ds = []
which_cases = []
for d in ds:
    min_case1 = (t/2) * d
    min_case2 = abs(d * int((B+1)/d) - (t/2) * d - 2)
    #min_case3
    if t %  d == 0:
        min_case3 = (t/d) * 2
    else:
        min_case3 = 2 * int((B+1)/d)
    
    mins = [min_case1, min_case2, min_case3]
    delta_d = min(mins)
    which_case = mins.index(delta_d)+1
    
    which_cases.append(which_case)
    delta_ds.append(delta_d)
    
d_border_to2 = ds[which_cases.index(2)]
print("d_border_to2 = ", d_border_to2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(20, 10) )
plt.plot(ds, delta_ds, marker='o')

plt.xticks(fontsize = 40, fontname="Times New Roman")
plt.yticks(fontsize = 40, fontname="Times New Roman")
plt.grid(True, which='major', linestyle='--' )
plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
plt.minorticks_on()

plt.xlabel('d', style='italic', fontname="Times New Roman", fontsize = 40)
plt.ylabel(r'$\delta(d)$', style='italic', fontname="Times New Roman", fontsize = 30)
```
