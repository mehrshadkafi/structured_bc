```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
```

```python
def get_histogram(image_name):
    '''Images_name should be like: 'lena', 'cam', 'man', 'liv', 'gol', or 'zelda' in tif
    format.'''
    img_file = image_name +'.tif'
    intensities_uint8 = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
    intensities = np.float32(intensities_uint8)

    num_rows = np.shape(intensities)[0]
    num_cols = np.shape(intensities)[1]

    diffs_rows = []
    for i in range(0, num_rows):
        diffs_row = []
        for j in range(0, num_cols-1):
            diff = abs(intensities[i][j+1] - intensities[i][j])
            diffs_row.append(diff)

        diffs_rows += diffs_row
        
    vals, counts = np.unique(diffs_rows, return_counts=True)
    prbs = counts/ sum(counts)
    val_count = tuple(zip(vals, counts))
    sort_by_val = sorted(val_count)

    count_val = tuple(zip(counts, vals))
    sort_by_count = sorted(count_val, reverse=True)

    plt.figure(figsize =(20, 10))
    plt.bar(vals, prbs)
    plt.xticks(fontsize = 40, fontname="Times New Roman")
    plt.yticks(fontsize = 40, fontname="Times New Roman")
    plt.grid(True, which='major', linestyle='--' )
    plt.grid(True, which='minor', color='#EEEEEE', linestyle='-' )
    plt.minorticks_on()
    plt.title (image_name, fontname="Times New Roman", fontsize = 40)

    return sort_by_count
```

```python
for image in ['lena', 'cam', 'man', 'liv', 'gol', 'zelda']:
    get_histogram(image)
```
